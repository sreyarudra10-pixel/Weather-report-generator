```python
import requests
```

```python
import os
```

```python
from datetime import datetime
```

```python
user_api = os.environ.get('OPENWEATHER_API_KEY')
location = input("Enter the city name: ")
```

```python
complete_api_link = "https://api.openweathermap.org/data/2.5/weather?q=" + location + "&appid=" + user_api
#This format was taken from the app which was given as https://api.openweathermap.org/data/2.5/weather?lat={lat}&lon={lon}&appid={API key} but we concatenated the location and key according to our convenience
```

```python
api_link = requests.get(complete_api_link) #requests.get() calls the API and returns a response object.
api_data = api_link.json()
```

```python
if api_data['cod'] == '404':  #Checks for a 404 error code in the JSON. #cod stands for “code” — it’s a key in the JSON response from the OpenWeatherMap API that tells the status of your request
    print("Invalid city: {}, please check the city name".format(location))
else:
    temp_city = ((api_data['main']['temp']) - 273.15)  #temp_city converts Kelvin → Celsius by subtracting 273.15 ,because the request didn’t specify units=metric
    weather_desc = api_data['weather'][0]['description']  #API returns list under 'weather'
    hmdt = api_data['main']['humidity'] #hmdt (humidity) and wind_spd (wind speed) are pulled from main and wind
    wind_spd = api_data['wind']['speed']
    date_time = datetime.now().strftime("%d %b %Y | %I:%M:%S %p")
```

```python
print("-----------------------------------------------------------")
print("weather stats for - {} || {}".format(location.upper(), date_time))  #exaple of format -> weather stats for - LONDON || 2026-01-21 08:00
print("-----------------------------------------------------------")
```

```python
print("Current temperature is : {:.2f}deg C".format(temp_city))
print("Current weather desc : ", weather_desc)
print("Current humidity : ", hmdt, '%')
print("Current wind speed : ", wind_spd, 'kmph')
```
